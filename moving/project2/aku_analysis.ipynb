```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
plt.style.use('ggplot')
import datetime as dt
import seaborn as sns
```

```python
aku = pd.read_csv("csvfile/aku.csv")
ari = pd.read_csv("csvfile/ari.csv")
ato = pd.read_csv("csvfile/ato.csv")
heart = pd.read_csv("csvfile/heart.csv")
sakai = pd.read_csv("csvfile/sakai.csv")
yamato = pd.read_csv("csvfile/yamato.csv")

aku = aku.drop(["Unnamed: 0"], axis=1)
ari = ari.drop(["Unnamed: 0"], axis=1)
ato = ato.drop(["Unnamed: 0"], axis=1)
heart = heart.drop(["Unnamed: 0"], axis=1)
sakai = sakai.drop(["Unnamed: 0"], axis=1)
yamato = yamato.drop(["Unnamed: 0"], axis=1)
```

# アーク引越し

```python
aku["引越し時期"] = pd.to_datetime(aku["引越し時期"])
```

```python
t1 = []
t2 = []
```

```python
for i in range(len(aku)):
   if aku["引越し先"][i] in ("青森", "岩手", "宮城", "秋田", "山形", "福島"):
      t1.append("東北")
   elif aku["引越し先"][i] in ("茨城", "栃木", "群馬", "埼玉", "千県", "東京", "神奈川", "山梨", "長野"):
      t1.append("関東")
   elif aku["引越し先"][i] in ("新潟", "富山", "石川",  "福井"):
      t1.append("北陸")
   elif aku["引越し先"][i] in ("岐阜",  "静岡",  "愛知",  "三重"):
      t1.append("中部")
   elif aku["引越し先"][i] in ("滋賀",  "京都",  "大阪",  "兵庫",  "奈良",  "和歌山"):
      t1.append("近畿")
   elif aku["引越し先"][i] in ("鳥取",  "島根",  "岡山",  "広島",  "山口"):
      t1.append("中国")
   elif aku["引越し先"][i] in ("徳島",  "香川",  "愛媛",  "高知"):
      t1.append("四国")
   elif aku["引越し先"][i] in ("福岡",  "佐賀",  "長崎",  "熊本",  "大分",  "宮崎",  "鹿児島"):
      t1.appned("九州")
   elif aku["引越し先"][i] == "沖縄":
      t1.append("沖縄")
   else:
      t1.append("北海道")
```

```python
for i in range(len(aku)):
   if aku["現在地"][i] in ("青森", "岩手", "宮城", "秋田", "山形", "福島"):
      t2.append("東北")
   elif aku["現在地"][i] in ("茨城", "栃木", "群馬", "埼玉", "千県", "東京", "神奈川", "山梨", "長野"):
      t2.append("関東")
   elif aku["現在地"][i] in ("新潟", "富山", "石川",  "福井"):
      t2.append("北陸")
   elif aku["現在地"][i] in ("岐阜",  "静岡",  "愛知",  "三重"):
      t2.append("中部")
   elif aku["現在地"][i] in ("滋賀",  "京都",  "大阪",  "兵庫",  "奈良",  "和歌山"):
      t2.append("近畿")
   elif aku["現在地"][i] in ("鳥取",  "島根",  "岡山",  "広島",  "山口"):
      t2.append("中国")
   elif aku["現在地"][i] in ("徳島",  "香川",  "愛媛",  "高知"):
      t2.append("四国")
   elif aku["現在地"][i] in ("福岡",  "佐賀",  "長崎",  "熊本",  "大分",  "宮崎",  "鹿児島"):
      t2.appned("九州")
   elif aku["現在地"][i] == "沖縄":
      t2.append("沖縄")
   else:
      t2.append("北海道")
```

```python
aku["移住地方"] = t2
aku["移住先地方"] = t1
```

```python
aku_2022 = aku[(aku["引越し時期"] > dt.datetime(2022,1,1)) & (aku["引越し時期"] < dt.datetime(2022,12,31))]
print(aku_2022.shape)
```
