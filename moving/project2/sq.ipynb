```python
import time
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
import datetime

date_data = []
compares_data = []
titles_data = []
scores_data = []
comments_data = []
current_place = []
moving_place = []
ppm_data = []
call_data = []
estimate_data = []
operator_data = []
service_data = []
kagu_data = []
kaden_data = []
danboll_data = []
compare_data = []

for n in range(1, 25):
  time.sleep(1)
  if n == 1:
    
    url = f'https://hikkoshizamurai.jp/company/d-the0123/voice/p1-com120'
  else:
      url = f'https://hikkoshizamurai.jp/company/d-the0123/voice/p1-com120?voice_page={n}'
  res = requests.get(url)
  soup = BeautifulSoup(res.text,"html.parser")

######################################################################
  # 日付の取得
  tables = soup.find_all("table",attrs={"voice-table"})
  for i in range(len(tables)):
    date = tables[i].text.replace("\n","").replace("引越し時期","").replace("年","-").replace("月","-")
    dt, _ = date.split("日(")
    date_data.append(dt)

  #　現在地と引越し先の都道府県の取得
  tables = soup.find_all("table",attrs={"voice-table"})
  for i in range(len(tables)):
    table1 = tables[i].text.replace("\n","").replace("引越し時期","").replace("年","-").replace("月","-").replace(" ","")
    table,pre = table1.split("住所")

    a,b = pre.split("→")
    if "移動距離" in b:
      b,_ = b.split("移動距離")
    elif "依頼作業開始時間" in b:
      b,_ = b.split("依頼作業開始時間")
    #b,_ = b.split("移動距離～")
    if  "大阪府" in a:
      a = "大阪"
    elif "京都府" in a:
      a = "京都"
    elif "北海道" in a:
      a = "北海道"
    elif "東京都" in a:
      a = "東京"
    elif "県" in a:
      a,_ = a.split("県")

    if  "大阪府" in b:
      b = "大阪"
    elif "京都府" in b:
      b = "京都"
    elif "北海道" in b:
      b = "北海道"
    elif "東京都" in b:
      b = "東京"
    elif "県" in b:
      b,_ = b.split("県")

    current_place.append(a)
    moving_place.append(b)

  # 依頼作業開始時間
  tables = soup.find_all("table",attrs={"voice-table"})
  for i in range(len(tables)):
    ppm = tables[0].text.replace("\n","").replace("引越し時期","").replace("年","-").replace("月","-").replace(" ","")
    _,ppm = ppm.split("依頼作業開始時間")
    ppm,_ = ppm.split("荷物量")
    ppm_data.append(ppm)

  #　電話・メール対応/見積もり対応/作業員/サービスの点数取得
  ls = soup.find_all("ul",attrs={"voice-rate-list"})
  for i in range(len(ls)):
    lsa = ls[i].text.replace("\n","").replace("電話・メール対応：","").replace("見積もり対応：","").replace("作業員：","").replace("サービス：","")
    call,estimate,operator,service = int(lsa[0]), int(lsa[1]), int(lsa[2]), int(lsa[3])
    call_data.append(call)
    estimate_data.append(estimate)
    operator_data.append(operator)
    service_data.append(service)

  #　荷物量の所得
  tables = soup.find_all("table",attrs={"voice-table"})
  for i in range(len(tables)):
    danball1 = tables[i].text.replace("\n","").replace("引越し時期","").replace("年","-").replace("月","-").replace(" ","")
    _,danball = danball1.split("依頼作業開始時間")
    danball,_ = danball.split("見積もり比較社数")

    if "ダンボール" in danball:
      danboll,num1 = danball.split("ダンボール")
      for i in range(len(num1)):
        if num1[i]=="/":
          num1 = int(num1[0:i])
          break
    else:
      num1 = 0
    danboll_data.append(num1)

    if "家具" in danball:
      danboll,num2 = danball.split("家具")
      for i in range(len(num2)):
        if num2[i]=="/":
          num2 = int(num2[0:i])
          break
    else:
      n = 0
    kagu_data.append(num2)


    if "家電" in danball:
      danboll,num3 = danball.split("家電")
      for i in range(len(num3)):
        if num3[i]=="/":
          num3 = int(num3[0:i])
          break
    else:
      num3 = 0
    kaden_data.append(num3)

    # 見積もり比較社数の取得
  tables = soup.find_all("table",attrs={"voice-table"})
  for i in range(len(tables)):
    estimate = tables[i].text.replace("\n","").replace("引越し時期","").replace("年","-").replace("月","-").replace(" ","")
    _,estimate = estimate.split("見積もり比較社数")
    estimate,m = estimate.split("訪問見積もり")
    if estimate == "-":
      compare = 0
    else:
      compare,estimate = estimate.split("社")
      compare = int(compare)
    compare_data.append(compare)


######################################################################

  # タイトルの取得
  titles = soup.find_all("div",attrs={"voice-title"})
  for t in titles:
    ts = t.text
    titles_data.append(ts)

  # 評価点の取得
  scores = soup.find_all("div",attrs={"voice-rate-number"})
  for s in scores:
    ss = s.text.replace("点","").replace("評価数：","").replace("\n", "").replace(" ","")
    scores_data.append(float(ss))

  # コメントの取得
  comments = soup.find_all("div",attrs={"voice-comment-text"})
  for c in comments:
    cs = c.text.replace("\n","").replace(" ","").replace("\r","")
    if len(cs) == 0:
      comments_data.append('none')
    else:
      comments_data.append(cs)



data = {}

data["引越し時期"] = date_data
data["タイトル"] = titles_data
data["評価"] = scores_data
data["口コミ"] = comments_data
data["現在地"] = current_place
data["引越し先"] = moving_place
data["依頼作業開始時間"] = ppm_data
data["電話・メール対応"] = call_data
data["見積もり対応"] = estimate_data
data["作業員"] = operator_data
data["サービス"] = service_data
data["家具量"] = kagu_data
data["家電量"] = kaden_data
data["ダンボール"] = danboll_data
data["見積もり比較社数"] = compare_data

df = pd.DataFrame(data)
```
